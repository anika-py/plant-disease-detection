## Binary Classification

```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the JSON file
with open("D:/VIT/Capstone/binaryclass model/binary_training_history_2.json", "r") as file:
    history_binary = json.load(file)

epochs = np.arange(0, len(history_binary["accuracy"]), step=1)

# Convert history to a Pandas DataFrame
df = pd.DataFrame(history_binary)
```

```python
# Function to add value annotations
def add_annotations(x, y):
    for i, txt in enumerate(y):
        plt.text(x[i], y[i], f"{txt:.2f}", ha="right", va="bottom", fontsize=9, color="black")
```

```python
# Plot Training vs Validation Accuracy
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["accuracy"], label="Training Accuracy", marker="o")
plt.plot(df.index, df["val_accuracy"], label="Validation Accuracy", marker="o")
add_annotations(df.index, df["accuracy"])
add_annotations(df.index, df["val_accuracy"])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.xticks(epochs)
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Training vs Validation Loss
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["loss"], label="Training Loss", marker="o")
plt.plot(df.index, df["val_loss"], label="Validation Loss", marker="o")
add_annotations(df.index, df["loss"])
add_annotations(df.index, df["val_loss"])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.xticks(epochs)
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Precision and Recall
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["precision"], label="Training Precision", marker="o")
plt.plot(df.index, df["val_precision"], label="Validation Precision", marker="o")
add_annotations(df.index, df["precision"])
add_annotations(df.index, df["val_precision"])
plt.xlabel("Epochs")
plt.ylabel("Score")
plt.xticks(epochs)
plt.title("Precision Trends")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["recall"], label="Training Recall", marker="o")
plt.plot(df.index, df["val_recall"], label="Validation Recall", marker="o")
add_annotations(df.index, df["recall"])
add_annotations(df.index, df["val_recall"])
plt.xlabel("Epochs")
plt.ylabel("Score")
plt.xticks(epochs)
plt.title("Recall Trends")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df["precision"], label="Training Precision", marker="o")
plt.plot(df.index, df["recall"], label="Training Recall", marker="o")
add_annotations(df.index, df["precision"])
add_annotations(df.index, df["recall"])
plt.xlabel("Epochs")
plt.ylabel("Score")
plt.xticks(epochs)
plt.title("Precision vs Recall Trends")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df["f1_score"] = 2 * (df["precision"] * df["recall"]) / (df["precision"] + df["recall"])
df["val_f1_score"] = 2 * (df["val_precision"] * df["val_recall"]) / (df["val_precision"] + df["val_recall"])

plt.figure(figsize=(10,5))
plt.plot(df.index, df["f1_score"], label="Training F1-Score", marker="o")
plt.plot(df.index, df["val_f1_score"], label="Validation F1-Score", marker="o")
add_annotations(df.index, df["f1_score"])
add_annotations(df.index, df["val_f1_score"])
plt.xlabel("Epochs")
plt.ylabel("F1 Score")
plt.xticks(epochs)
plt.title("Training vs Validation F1-Score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
learning_rates = history_multi["learning_rate"]
accuracy = history_multi["accuracy"]
val_accuracy = history_multi["val_accuracy"]
loss = history_multi["loss"]

# Function to plot with labels
def plot_with_labels(x, y, title, ylabel, color):
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, marker='o', linestyle='-', color=color)

    plt.xlabel("Learning Rate")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid()
    plt.show()

# Plot Learning Rate vs Accuracy
plot_with_labels(learning_rates, accuracy, "Learning Rate vs Accuracy", "Accuracy", "b")

# Plot Learning Rate vs Loss
plot_with_labels(learning_rates, loss, "Learning Rate vs Loss", "Loss", "r")

# Plot Learning Rate vs Validation Accuracy
plot_with_labels(learning_rates, val_accuracy, "Learning Rate vs Validation Accuracy", "Validation Accuracy", "g")
```

```python
learning_rates = history_multi["learning_rate"]
epochs = range(1, len(history_multi["accuracy"]) + 1)
accuracy = history_multi["accuracy"]
val_accuracy = history_multi["val_accuracy"]
loss = history_multi["loss"]

def plot_epochs_vs_metric(epochs, metric_values, title, ylabel, color, learning_rates):
    plt.figure(figsize=(8, 5))
    
    unique_lrs = sorted(set(learning_rates))  # Sort for better visualization

    for lr in unique_lrs:
        # Extract indices where the learning rate matches
        indices = [i for i in range(len(learning_rates)) if learning_rates[i] == lr]

        # Get corresponding epochs and metric values
        epochs_subset = [epochs[i] for i in indices]
        metric_subset = [metric_values[i] for i in indices]

        # Ensure epochs and metric values have the same length
        if len(epochs_subset) == len(metric_subset):
            plt.plot(epochs_subset, metric_subset, marker='o', linestyle='-', label=f'LR={lr:.6f}', alpha=0.8)
        else:
            print(f"Warning: Mismatched lengths for LR={lr}")

    plt.xlabel("Epochs")
    plt.ylabel(ylabel)
    plt.xticks(epochs_m+1)
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Plot Epochs vs Accuracy for different Learning Rates
plot_epochs_vs_metric(epochs, accuracy, "Epochs vs Accuracy for Different Learning Rates", "Accuracy", "b", learning_rates)

# Plot Epochs vs Loss for different Learning Rates
plot_epochs_vs_metric(epochs, loss, "Epochs vs Loss for Different Learning Rates", "Loss", "r", learning_rates)

# Plot Epochs vs Validation Accuracy for different Learning Rates
plot_epochs_vs_metric(epochs, val_accuracy, "Epochs vs Validation Accuracy for Different Learning Rates", "Validation Accuracy", "g", learning_rates)
```

```python
from sklearn.metrics import f1_score

def calculate_f1(precision, recall):
    return [(2 * p * r) / (p + r) if (p + r) != 0 else 0 for p, r in zip(precision, recall)]

def tabulate_metrics(data, classification_type):
    df = pd.DataFrame({
        "Epoch": list(range(1, len(data["accuracy"]) + 1)),
        "Accuracy": data["accuracy"],
        "Precision": data["precision"],
        "Recall": data["recall"],
        "F1-Score": calculate_f1(data["precision"], data["recall"]),
    })
    if "learning_rate" in data:
        df["Learning Rate"] = data["learning_rate"]
    print(f"\n{classification_type} Classification Metrics:\n")
    print(df.to_string(index=False))

tabulate_metrics(history_binary, "Binary")
tabulate_metrics(history_multi, "Multiclass")
```
